## --- Numpy ---

```python
# Importamos la librería numpy, fundamental para el cálculo numérico y manejo de arrays en Python.
import numpy as np

# Creamos un array bidimensional (matriz) de 2 filas y 3 columnas con valores específicos.
x = np.array([[1, 2, 3], [4, 5, 6]])
# Imprimimos el array 'x' usando formato de cadena para visualizarlo de forma ordenada.
print("x:\n{}".format(x))
```

## --- Scipy ---

```python
# Importamos el submódulo 'sparse' de scipy, que nos permite trabajar con matrices dispersas,
# es decir, matrices que tienen muchos ceros y se pueden almacenar de manera eficiente.
from scipy import sparse

# Creamos una matriz identidad de 4x4 usando numpy.
# Una matriz identidad tiene 1's en la diagonal principal y 0's en el resto.
eye = np.eye(4)
# Imprimimos la matriz identidad creada, mostrando que es un array de numpy.
print("Numpy array:\n{}".format(eye))
```

```python
# Convertimos la matriz identidad en una matriz dispersa en formato CSR (Compressed Sparse Row).
# Este formato es útil para operaciones aritméticas y de álgebra lineal en matrices dispersas.
sparse_matrix = sparse.csr_matrix(eye)
# Imprimimos la representación CSR de la matriz, en la cual se almacenan sólo los elementos no nulos.
print("\nScipy sparse CSR matrix:\n{}".format(sparse_matrix))
```

```python
# Creamos un vector de datos que contiene 4 elementos, todos iguales a 1.
# Estos serán los valores no nulos que se colocarán en la matriz dispersa.
data = np.ones(4)

# Creamos un array que contiene los índices de fila para cada uno de los elementos de 'data'.
# En este caso, generamos un array [0, 1, 2, 3] usando np.arange, lo cual significa que cada 1
# se colocará en la diagonal de la matriz.
row_indices = np.arange(4)

# De manera similar, creamos un array para los índices de columna, también [0, 1, 2, 3].
col_indices = np.arange(4)

# Creamos una matriz dispersa en formato COO (Coordinate format) usando los arrays 'data',
# 'row_indices' y 'col_indices'. Este formato almacena la posición (fila, columna) de cada elemento.
eye_col = sparse.coo_matrix((data, (row_indices, col_indices)))
# Imprimimos la representación COO de la matriz, mostrando cómo se almacena la información.
print("COO representation:\n{}".format(eye_col))
```

## --- Seaborn y Matplotlib para Visualización ---

```python
# Importamos seaborn para mejorar la estética de los gráficos y matplotlib.pyplot para funciones adicionales.
import seaborn as sns
import matplotlib.pyplot as plt

# Establecemos el estilo de Seaborn. "whitegrid" añade una cuadrícula blanca que facilita la lectura del gráfico.
sns.set_theme(style="whitegrid")

# Generamos 100 valores equidistantes entre -10 y 10 para el eje x.
x = np.linspace(-10, 10, 100)

# Calculamos el valor del seno para cada valor de x.
y = np.sin(x)

# Creamos un gráfico de línea usando Seaborn. El parámetro marker="x" marca cada punto con una 'x'.
sns.lineplot(x=x, y=y, marker="x")

# Añadimos título y etiquetas a los ejes para hacer el gráfico más descriptivo.
plt.title("Función Seno")
plt.xlabel("Valores de x")
plt.ylabel("Seno de x")

# Mostramos el gráfico. En un Jupyter Notebook, plt.show() generalmente se invoca de manera automática,
# pero se recomienda incluirlo para asegurar la visualización del gráfico.
plt.show()
```
